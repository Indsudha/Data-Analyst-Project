```python
mean = 5
std_dev = 1.5
```

##2.1 What proportion of the gunny bags have a breaking strength of less than 3.17 kg per sq cm?

```python
def plot_normal_dist(x_values, condition):
  x = np.linspace(mean - 3 * std_dev, mean + 3 * std_dev, 1000)
  plt.plot(x, stats.norm.pdf(x, mean, std_dev))
  plt.title("Normal Distribution")

  if condition == "less_than":
    plt.fill_between(x, stats.norm.pdf(x, mean, std_dev), where=x < x_values[0], color='gray', alpha=0.5)
  elif condition == "at_least":
    plt.fill_between(x, stats.norm.pdf(x, mean, std_dev), where=x >= x_values[0], color='gray', alpha=0.5)
  elif condition == "between":
    plt.fill_between(x, stats.norm.pdf(x, mean, std_dev), where=(x > x_values[0]) & (x < x_values[1]), color='gray', alpha=0.5)
  elif condition == "not_between":
    plt.fill_between(x, stats.norm.pdf(x, mean, std_dev), where=(x < x_values[0]) | (x > x_values[1]), color='gray', alpha=0.5)

  plt.show()
```

```python
x = 3.17
proportion = stats.norm.cdf(x, loc=mean, scale=std_dev)
print(f"Proportion of bags with breaking strength less than {x}: {proportion}")
plot_normal_dist([x], "less_than")
```

###State the null and alternate hypotheses

Null Hypothesis
H₀ (Null): Mean hardness of unpolished stones ≥ 150

H₁ (Alternative): Mean hardness of unpolished stones < 150


```python
columns_list = df.columns.tolist()
print(columns_list)
```

```python
unpolished = df['Unpolished '].dropna()
polished = df['Treated and Polished'].dropna()
```

```python
UP_mean = df['Unpolished '].mean()
UP_std_dev = df['Unpolished '].std()

print("Mean:", UP_mean)
print("Standard Deviation:", UP_std_dev)
```

###Conduct the hypothesis test and compute the p-value

```python
# 3.1 - One-sample t-test: Is the mean of unpolished stones < 150?

t_stat_1, p_value_1 = stats.ttest_1samp(unpolished, 150)
p_value_1_one_tailed = p_value_1 / 2  # since it's a one-tailed test
```

###Write down conclusions from the test results

```python
# Print results
print("=== 3.1: One-Sample T-Test (Unpolished vs 150) ===")
print(f"Mean (Unpolished): {unpolished.mean():.2f}")
print(f"T-statistic: {t_stat_1:.4f}")
print(f"One-tailed p-value: {p_value_1_one_tailed:.5f}")
```

```python
if p_value_1_one_tailed < 0.05:
    print("Conclusion: Reject the null hypothesis. The unpolished stones likely have hardness < 150.")
else:
    print("Conclusion: Fail to reject the null. There's not enough evidence to say hardness < 150.")
```

####Test Result

Sample Mean (Unpolished): 134.11

One-tailed p-value: 0.00004

Conclusion (α = 0.05):
Since the p-value is much less than 0.05, we reject the null hypothesis.

Yes, Zingaro is justified in believing that the unpolished stones may not be suitable for printing, as their average hardness is significantly below 150.

```python
# 3.2 - Two-sample t-test: Are the means of unpolished and polished stones equal?

t_stat_2, p_value_2 = stats.ttest_ind(unpolished, polished, equal_var=False)
```

###Write down conclusions from the test results.

```python
print("\n=== 3.2: Two-Sample T-Test (Unpolished vs Polished) ===")
print(f"Mean (Polished): {polished.mean():.2f}")
print(f"Mean (Unpolished): {unpolished.mean():.2f}")
print(f"T-statistic: {t_stat_2:.4f}")
print(f"Two-tailed p-value: {p_value_2:.5f}")
```

```python
if p_value_2 < 0.05:
    print("Conclusion: Reject the null hypothesis. The means are significantly different.")
else:
    print("Conclusion: Fail to reject the null. The means are not significantly different.")
```

###Conduct the hypothesis test and compute the p-value

```python
df['Alloy'] = df['Alloy'].astype(int).map({1: 'Alloy1', 2: 'Alloy2'}).astype('category')
```

```python
# Perform one-way ANOVA for Alloy-1
fvalue1, pvalue1 = stats.f_oneway(
    df_alloy1['Response'][df_alloy1['Dentist'] == 1],
    df_alloy1['Response'][df_alloy1['Dentist'] == 2],
    df_alloy1['Response'][df_alloy1['Dentist'] == 3],
    df_alloy1['Response'][df_alloy1['Dentist'] == 4],
    df_alloy1['Response'][df_alloy1['Dentist'] == 5]
)

print("Alloy-1 ANOVA results:")
print(f"F-value: {fvalue1}")
print(f"P-value: {pvalue1}")

# Perform one-way ANOVA for Alloy-2
fvalue2, pvalue2 = stats.f_oneway(
    df_alloy2['Response'][df_alloy2['Dentist'] == 1],
    df_alloy2['Response'][df_alloy2['Dentist'] == 2],
    df_alloy2['Response'][df_alloy2['Dentist'] == 3],
    df_alloy2['Response'][df_alloy2['Dentist'] == 4],
    df_alloy2['Response'][df_alloy2['Dentist'] == 5]
)

print("\nAlloy-2 ANOVA results:")
print(f"F-value: {fvalue2}")
print(f"P-value: {pvalue2}")
```

###Write down conclusions from the test results

For both Alloy-1 and Alloy-2, the p-values are greater than the significance level of 0.05. This means we fail to reject the null hypothesis for both alloys.

Therefore, we conclude that there is no statistically significant difference in the mean hardness values between the different dentist groups for either Alloy-1 or Alloy-2.

In simpler terms, the hardness of implants does not appear to be significantly influenced by the dentist performing the assessment for either type of alloy.

###State the null and alternate hypotheses

####Hypotheses (Two-Way ANOVA with Interaction)

Null Hypotheses (H₀):

H₀₁: There is no difference in implant hardness across different methods.

H₀₂: There is no difference in implant hardness across different dentists.

H₀₃: There is no interaction effect between dentist and method on implant hardness.

Alternative Hypotheses (H₁):

H₁₁: At least one method differs in mean hardness.

H₁₂: At least one dentist differs in mean hardness.

H₁₃: There is an interaction effect between dentist and method.

####Fit the Two-Way ANOVA Model

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Model for Alloy-1
model_alloy1 = ols('Response ~ C(Method) + C(Dentist) + C(Method):C(Dentist)', data=df_alloy1).fit()

# Model for Alloy-2
model_alloy2 = ols('Response ~ C(Method) + C(Dentist) + C(Method):C(Dentist)', data=df_alloy2).fit()
```

###Check the assumptions of the hypothesis test



Normality: Use Shapiro-Wilk test for residuals.

Homogeneity of Variance: Use Levene’s test. (Note: Even if assumptions fail, proceed as instructed.)



####1. Shapiro-Wilk Test (Normality of Residuals)

```python
from scipy.stats import shapiro

# Residuals for Alloy-1
residuals1 = model_alloy1.resid
shapiro_test1 = shapiro(residuals1)
print("Alloy-1 Shapiro-Wilk Test (Normality):")
print(f"Statistic = {shapiro_test1.statistic:.4f}, p-value = {shapiro_test1.pvalue:.4f}")
if shapiro_test1.pvalue > 0.05:
    print("Residuals are approximately normal.\n")
else:
    print("Residuals are not normal.\n")
```

Normality Assumption Check (Shapiro-Wilk Test) – Alloy-1

The Shapiro-Wilk test for the residuals of Alloy-1 yielded a test statistic of 0.9565 and a p-value of 0.0902.

Since the **p-value is greater than the significance level of 0.05**, we fail to reject the null hypothesis of the test. This indicates that the residuals are approximately **normally distributed**, satisfying the assumption of normality for ANOVA.

```python
# Residuals for Alloy-2

residuals2 = model_alloy2.resid
shapiro_test2 = shapiro(residuals2)
print("Alloy-2 Shapiro-Wilk Test (Normality):")
print(f"Statistic = {shapiro_test2.statistic:.4f}, p-value = {shapiro_test2.pvalue:.4f}")
if shapiro_test2.pvalue > 0.05:
    print("Residuals are approximately normal.\n")
else:
    print("Residuals are not normal.\n")
```

Normality Assumption Check (Shapiro-Wilk Test) – Alloy-2

The Shapiro-Wilk test for the residuals of Alloy-2 produced a test statistic of 0.9571 and a p-value of 0.0946.

As the **p-value exceeds the standard significance level of 0.05**, we fail to reject the null hypothesis of normality.
This suggests that the residuals are approximately **normally distributed**, thereby satisfying the assumption of normality required for conducting ANOVA.
